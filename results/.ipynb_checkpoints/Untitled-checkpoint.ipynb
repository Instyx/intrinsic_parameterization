```python
meshes = df['mesh name'].unique()
df['flipping order 2']=[df.loc[i+1,'flipping order'] if x=='start' else x for i,x in enumerate(df['flipping order']) ]
types = df['type'].unique()
flipping_orders = df['flipping order 2'].unique()
boundary = df['free boundary'].unique()
# Specify the flipping order to exclude
exclude_order = 'start'


for mesh in meshes:
    for t in types:
        for b in boundary:
            for order in flipping_orders:
                df_subset = df[(df['mesh name'] == mesh) & (df['type'] == t) & (df['flipping order'] == order) & df['free boundary'] == b]
                display(df_subset.head(n=6))
                if len(df_subset)!=0:
                    fig=sns.barplot(data=df_subset, x="flip iteration", y="energy", hue="flipping order 2")
                    display(fig)
```
