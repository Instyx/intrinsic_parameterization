```python
def sns_viz(data, palet = "Set2", style="whitegrid"):
    plt.figure(figsize=(6.5, 3.25))
    
    sns.set_theme(style=style)
    fig = sns.lineplot(data=data, x="Iteration", y="Energy", hue='Intri', style='Intri', palette='deep')
    fig = sns.scatterplot(data=data, x="Iteration", y="Energy", marker="o", hue='Type', palette=palet)
    
    fig.set_xlabel('Iteration')
    fig.set_ylabel('Energy / Total Area')
    fig.set_title(f"Global/Local Intrinsic Param (Octopus (small), Free, Sym. Dirichlet)")
    _, ax = plt.subplots()
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles=handles[1:], labels=labels[1:])
    plt.savefig('octo_slim_iparam.svg', format='svg')
    plt.show()
```
