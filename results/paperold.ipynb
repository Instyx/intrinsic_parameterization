```python
def sns_viz(data, mesh, palet = "dark", style="whitegrid"):
    plt.figure(figsize=(6.5, 3.25))
#     new_list = range(math.floor(min(data['Iteration'])), math.ceil(data['Iteration'])+1)
#     plt.xticks(new_list)
    palet_used = sns.color_palette(palet, n_colors=8)
    palet_used.reverse()
    fig, ax = plt.subplots()
    ax.xaxis.set_major_locator(MaxNLocator(integer=True))
    sns.set(font=prop.get_name(), style=style)
    g = sns.scatterplot(data=data, x="Iteration", y="Energy", marker="o", s=60, hue='Type', palette=palet_used)

    g = sns.lineplot(data=data, x="Iteration", y="Energy", hue='Method', palette='husl', alpha=0.4)


    g.set_xlabel('Iteration')
    g.set_ylabel('Normalized Energy')
    #g.set_title(f"{mesh}")
    legend = ax.legend()
#     handles, labels = ax.get_legend_handles_labels()
#     ax.legend(handles=handles[1:], labels=labels[1:])  
    plt.savefig(f"{mesh}.svg", format='svg')
    plt.show()
```

```python
# Specify the directory path
directory = '/home/korau/output_iparam_queue'
#directory = '../output'
# Iterate over the folders in the directory
res = []
for folder in os.listdir(directory):
    # Check if the item is a folder
    mesh_folder = os.path.join(directory, folder)
    all_energies = []
    all_iterations = []
    all_types = []
    all_methods = []
    for item in os.listdir(mesh_folder):
        log_path = os.path.join(mesh_folder, item)
        if os.path.isdir(log_path):
            continue
        if os.path.getsize(log_path) == 0:
            continue
        with open(log_path, 'r') as file:
            # Create a CSV reader object
            reader = csv.reader(file, delimiter=',')
            #print(file)
            first_row = next(reader)
            method_name = first_row[-1]
            if(method_name == 'arap'):
                method_name = 'ARAP'
            if(method_name == 'asap'):
                method_name ='ASAP'
            if(method_name == 'dirichlet'):
                method_name = 'Dirichlet'
            if(method_name == 'symdirichlet'):
                method_name = 'S. Dirichlet'
                                
            name_m = first_row[0]
            # Iterate over each row in the log file

            for row in reader:
                if row[0] == 'energies':
                    energies_str = row[1:]
                    energies = [float(i) for i in energies_str]
                    max_energy = max(energies)
                    energies_normalized = [i/max_energy for i in energies]
                    iterations = [i for i,_ in enumerate(energies)]
                    types = ['Position Opt.' if i%2==0 else 'Flipping' for i in iterations]
                    iterations_half = [i/2 for i in iterations]
                    methods = [method_name for i in iterations]
                    all_energies += energies_normalized
                    all_iterations += iterations_half
                    all_types += types
                    all_methods += methods
    
    if all_methods==[]:
        continue   
    data_curr = { 'Energy' : all_energies,
                      'Iteration' : all_iterations,
                      'Type' : all_types,
                      'Method': all_methods
                    }
    if name_m == '106620.off': 
        df = pd.DataFrame(data_curr)
        sns_viz(df, name_m)
```
